0. Importamos librerias necesarias

```python
#importamos pandas y matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

1.  Convierte los archivos funds.csv y funding_rounds.csv a pandas dataframes.

```python
#se usa read_CSV para abrir los archivos como pandas frames
datfra1=pd.read_csv(r'C:\Users\yonat\OneDrive\Documents\IMF UCAV Master Data Science\M1 Las herramientas del científico de datos (presencial 042024)\Práctica Final Python\funds.csv')
datfra2=pd.read_csv(r'C:\Users\yonat\OneDrive\Documents\IMF UCAV Master Data Science\M1 Las herramientas del científico de datos (presencial 042024)\Práctica Final Python\funding_rounds.csv')
```

2.  Busca un elemento común de ambos dataframes y conviértelos en uno solo.

```python
#se usa concat para compilar los data frames con inner para combinar los data frames a uno solo
concatdf=pd.concat([datfra1,datfra2],ignore_index=True, join='inner')
concatdf
```

3.  Elimina las columnas 'id', 'source_url', 'source_description', 'updated_at'.

```python
#se definen las columnas a quitar
columnas_a_quitar = ['id', 'source_url','source_description', 'updated_at']
#las quitamos con drop
concatdf.drop (columnas_a_quitar, axis=1, inplace=True)
concatdf
```

4. Muestra el número de nulos que tiene cada columna del dataframe y qué porcentaje supone dentro de cada columna el número de nulos.

```python
# Calcula el número de nulos en cada columna
nancount = concatdf.isnull().sum()  
# Calcula el total de filas en el DataFrame con len
total_filas = len(concatdf)  
# Calcula el porcentaje de nulos en cada columna
nan_porcentaje = (nancount / total_filas) * 100  
# se utilizan diferentes nombres de variables para los valores calculados dentro del bucle
for columna, nulos, porcentaje in zip(nancount.index, nancount, nan_porcentaje):
   print(f'Columna "{columna}": {nulos} nulos, {porcentaje:.2f}%')
```

5.  Modifica todos los valores de 'raised_currency_dode' nulos a 'USD'

```python
#remplazamos los nulos a USD
concatdf['raised_currency_code'].fillna('USD', inplace=True)
concatdf
```

6. Elimina el resto de nulos del dataset.

```python
#eliminamos los nulos
concatdf.dropna(inplace=True)
concatdf
```

```python
#verificamos que se hayan borrado
concatdf.isnull().sum()
```

7.  Muestra el porcentaje de los diferentes valores de 'raised_currency_code'.

```python
#contamos los valores y sacamos su distribución para multiplicarlo por 100 y damos formato para que aparezca en porcentajes y con 2 decimales
currency_distribution = concatdf['raised_currency_code'].value_counts(normalize=True) * 100
currency_distribution = currency_distribution.apply(lambda x: f'{x:.2f}%')
print(currency_distribution)
```

8.  Crea un sub-conjunto sin el valor USD y visualiza un diagrama de barras.

```python
#quitamos todo lo que sea distinto a USD y lo proyectamos en forma de diagrama de barras con titulo, y etiquetas en los ejes
sin_usd_df = concatdf[concatdf['raised_currency_code'] != 'USD']
sin_usd_currency_counts = sin_usd_df['raised_currency_code'].value_counts()
sin_usd_currency_counts.plot(kind='bar')
plt.title('Distribución de las monedas que no son USD')
plt.xlabel('Código de Moneda')
plt.ylabel('Frecuencia')
plt.show()
```

9.  Crea un nuevo sub-conjunto filtrando del dataset todas las filas por debajo de 10M y visualiza el histograma.

```python
#Quitamos todo lo menor a 10M para graficarlo en forma de histograma
less10Mfunding_df = concatdf[concatdf['raised_amount'] >= 10000000]
less10Mfunding_df['raised_amount'].plot(kind='hist', bins=50)
plt.title('Distribución de los montos (>= 10M)')
plt.xlabel('Monto de Ingreso')
plt.ylabel('Frecuencia')
plt.show()
```

10.  Con el conjunto del apartado anterior. Investiga sobre la función "contains()" filtra el dataset por las fechas de fundación de las startup en 2000

```python
# Convertimos la columna 'funded_at' a string y luego extraemos el año para comparar y averiguamos el tipo de resultado
created_in_2000_df = less10Mfunding_df['funded_at'].astype(str).str.contains('2000', regex=False)
#created_in_2000_df
print(type(created_in_2000_df))
```

```python
#filtramos la serie por cuando fue fundada
less10Mfunding_df = less10Mfunding_df[created_in_2000_df].sort_values(by='funded_at', ascending= True)
less10Mfunding_df
```

11.  Sobre el conjunto filtrado por fecha visualiza la cuantía obtenida explicada por el tipo de moneda.

```python
amount_by_currency = less10Mfunding_df.groupby('raised_currency_code')['raised_amount'].sum()
amount_by_currency.plot(kind='bar')
plt.title('Total Raised Amount in 2000 by Currency Code')
plt.xlabel('Currency Code')
plt.ylabel('Total Raised Amount (USD)')
plt.show()
```
